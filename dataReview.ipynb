```python
import numpy as np
import pandas as pd
```

```python
files = !ls dataExamples
files = [file for file in files if '.npy' in file and 'vibration' in file]
```

```python
summary = []
for file in files:
    data = np.load('/'.join(['dataExamples', file]))
    diffData = np.diff(data)
    v, n = np.unique(diffData, return_counts=True)
    p = n / diffData.size
    
    v0, n0 = np.unique(data, return_counts=True)
    p0 = n0 / data.size
    summary.append({
        'file': file,
        'entropy after diff': np.sum(-p * np.log2(p)), 
        'entropy': np.sum(-p0 * np.log2(p0)), 
        'nValues in dict': n0.size,
        'nValues in dict after diff': n.size,
        'Signal length': data.size
    })
```

```python
pd.DataFrame(summary).sort_values(by='file')
```

```python
pd.DataFrame(summary).sort_values(by='file').mean()
```

```python
pd.DataFrame(summary).sort_values(by='file').std()
```
